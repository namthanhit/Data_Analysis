## Data crawl từ batdongsan.com

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import undetected_chromedriver as uc
import time
import pandas as pd
```

# 1 - 200

```python
# ==== CẤU HÌNH SELENIUM ====
options = uc.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-notifications")
options.add_argument("--disable-popup-blocking")

# Khởi tạo trình duyệt Chrome
driver = uc.Chrome(options=options, version_main=136)

# ==== CRAWL NHIỀU TRANG ====
base_url = "https://batdongsan.com.vn/nha-dat-ban-ha-noi/p"
all_links = []
start_page = 1  # Bắt đầu từ trang 1
max_pages = 200  # Tuỳ chỉnh số trang muốn crawl

for page in range(start_page, max_pages + 1):
    url = f"{base_url}{page}"
    print(f"Đang xử lý trang {page}: {url}")

    try:
        driver.get(url)
        time.sleep(3)  # Chờ trang load

        # Scroll để đảm bảo tất cả bài đăng load ra
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)


        # Lấy tất cả thẻ chứa link bài đăng
        listings = driver.find_elements(By.CLASS_NAME, "js__product-link-for-product-id")

        for item in listings:
            link = item.get_attribute("href")
            if link and link.startswith("https://batdongsan.com.vn/") and "booking" not in link:
                all_links.append(link)

    except Exception as e:
        print(f"Lỗi ở trang {page}: {e}")
        continue

# Đóng trình duyệt
driver.quit()

# ==== LƯU KẾT QUẢ ====
df_links = pd.DataFrame(all_links, columns=["Link"])
df_links.drop_duplicates(inplace=True)
df_links.to_csv("Link\\links_bds_hanoi_1.csv", index=False, encoding="utf-8-sig")

print(f"Hoàn tất. Đã lấy được {len(df_links)} link và lưu vào 'links_bds_hanoi_1.csv'")
```
